```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../cleaned_data.csv')
```

```python
plot = sns.barplot(data=df['play_type'], errorbar=None)
plot.set_title("Play Type Distribution")
plt.show()
```

```python
plot = sns.histplot(data=df, x='yardline_100')
plot.set_title("Yardline 100 Distribution")
plt.show()
```

```python
plot = sns.histplot(data=df, x='wp')
plot.set_title("Win Percentage Distribution")
plt.show()
```

```python
plot = sns.histplot(data=df, x='down', hue='play_type', multiple='stack', discrete=True)
plot.set_title("Play Type by Down")
plt.xticks([1, 2, 3, 4]) 
plt.show()
```

```python
plot = sns.histplot(data=df, x='game_seconds_remaining', hue='play_type', multiple='stack', bins=12)
plot.set_title("Play Type by Game Seconds Remaining")
plt.xticks([_ for _ in range(0, 3900, 300)]) 
plt.gca().invert_xaxis()
plt.show()
```
