```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../cleaned_data.csv')
```

```python
plot = sns.histplot(data=df, x='play_type')
plot.set_title("Play Type Distribution")
plt.show()
```

```python
plot = sns.histplot(data=df, x='yardline_100', binwidth=1)
plot.set_title("Yardline 100 Distribution")
plot.set_xticks(range(0, 101, 10))
plt.show()
```

```python
plot = sns.histplot(data=df, x='wp', hue='play_type', multiple='stack', bins=20)
plot.set_title("Play Type by Win Percentage Distribution")
plt.show()
```

```python
plot = sns.histplot(data=df, x='down', hue='play_type', multiple='stack', discrete=True)
plot.set_title("Play Type by Down")
plt.xticks([1, 2, 3, 4]) 
plt.show()
```

```python
plot = sns.histplot(data=df, x='game_seconds_remaining', hue='play_type', multiple='stack', bins=12)
plot.set_title("Play Type by Game Seconds Remaining")
plt.xticks(range(0, 3601, 300))
plt.gca().invert_xaxis()
plt.show()
```

```python
plot = sns.histplot(data=df, x='roof', hue='play_type', multiple='stack')
plot.set_title("Play Type by Roof")
plt.xticks(['outdoors', 'open', 'closed', 'dome'])
plt.show()
```
